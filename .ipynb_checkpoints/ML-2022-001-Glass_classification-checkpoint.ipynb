# 1 - Introduction <a class="anchor" id="chapter1"></a>

The purpose of this study is to define the type of a glass depending on the given features. 6 types of glass; defined in terms of their oxide content (i.e. Na, Fe, K, etc) and refractive index (RI). There are originally seven type of glasses but type-4 is omitted from the data set hence there are six type of glass in the data set.

The motivation behind the glass classification is given by the contributor as;
>The study of classification of types of glass was motivated by
criminological investigation. At the scene of the crime, the glass left
can be used as evidenceâ€¦if it is correctly identified!

The data set used in this work is obtained from;
- https://archive.ics.uci.edu/ml/datasets/Glass+Identification
- https://www.kaggle.com/datasets/uciml/glass

Columns names and also the ingredients are:
- RI: refractive index
- Na: Sodium (unit measurement: weight percent in corresponding oxide, as are attributes 4-10)
- Mg: Magnesium
- Al: Aluminum
- Si: Silicon
- K: Potassium
- Ca: Calcium
- Ba: Barium
- Fe: Iron
- Type of glass: (class attribute)
    - 1 building*windows*float*processed*
    - 2 building*windows*non*float*processed
    - 3 vehicle*windows*float*processed*
    - ~~4 vehicle*windows*non*float*processed~~ (none in this database)
    - 5 containers
    - 6 tableware
    - 7 headlamps

Questions by the contributor:
>Data exploration of this dataset reveals two important characteristics :
>1) The variables are highly corelated with each other including the response variables:
>So which kind of ML algorithm is most suitable for this dataset Random Forest , KNN or other? Also since dataset is too small is there any chance of applying PCA or it should be completely avoided?
>
>2) Highly Skewed Data:
>Is scaling sufficient or are there any other techniques which should be applied to normalize data? Like BOX-COX Power transformation?

Let's check the mutual information. Code below is taken and editted from Kaggle's [mutual information](https://www.kaggle.com/code/ryanholbrook/mutual-information) course.

```python
def make_mi_scores(X, y, discrete_features):
    mi_scores = mutual_info_classif(X, y, discrete_features=discrete_features)
    mi_scores = pd.Series(mi_scores, name="MI Scores", index=X.columns)
    mi_scores = mi_scores.sort_values(ascending=False)
    return mi_scores

mi_scores = make_mi_scores(glass[features], glass['Type'], 'auto')
mi_scores#[::3]  # show a few features with their MI scores
```

## 3.1 - Droping Outliers

```python
glass_outs_dropped = glass.copy()
glass_outs_dropped.drop(index = [105,106,111,162,170,171,183,200], 
                       axis = 0, 
                       inplace = True
                      )
glass_outs_dropped.iloc[103:113]
```

## 3.2 - Scaling Oxide Contents

The purpose of the cell below is to set the oxide contents between 0-1 instead of 0-100.

```python
glass_RI = glass_outs_dropped.iloc[:,:1].copy() # Copy 'RI' column w/o scaling.
glass_Type = glass_outs_dropped.iloc[:,-1:].copy() # Copy 'Type' column w/o scaling.
    
for feature in features[1:]: # Scaling other features b/w 0-1.
    df_scaled = glass_outs_dropped.iloc[:,1:9].div(100)#.round(2)
    
glass_oc_scaled = pd.concat([glass_RI.reset_index(drop=True), 
                          df_scaled.reset_index(drop=True), 
                          glass_Type.reset_index(drop=True)], 
                         axis=1)
glass_oc_scaled
```

## 3.3 - Scaling RI (MinMax Scaling)

```python
glass_RI_scaled = minmax_scaling(glass_RI, columns=['RI']) # Minmax scale RI column.

glass_scaled_wo_RI = glass_oc_scaled.iloc[:,1:].copy() # Copy 'RI' column w/o scaling.

glass_all_scaled = pd.concat([glass_RI_scaled.reset_index(drop=True), 
                              glass_scaled_wo_RI.reset_index(drop=True)], 
                             axis=1)
glass_all_scaled
```

## 3.4 - Normalisation

Until now we obtained four data sets, three of them are altered, `glass`, `glass_outs_dropped`, `glass_oc_scaled`, and `glass_all_scaled`. Let's obtain normalised versions of those data sets.

```python
def normaliser(X, features):
    X_norm = []
    
    for feature in features:
        X_bc_transformed, _ = stats.boxcox(X[feature]+1)
        
        X_norm.append(X_bc_transformed)
    X_norm = np.column_stack(X_norm)
    X_norm = pd.DataFrame(data=X_norm, columns=features)
    X_norm = pd.concat([X_norm.reset_index(drop=True), 
                        X['Type'].reset_index(drop=True)], 
                       axis=1)
    
    return X_norm

glass_norm = normaliser(glass, features)
gls_ots_drp_norm = normaliser(glass_outs_dropped, features)
gls_oc_scl_norm = normaliser(glass_oc_scaled, features)
gls_all_scl_norm = normaliser(glass_all_scaled, features)
```

# 4 - Models <a class="anchor" id="chapter4"></a>

## 4.1 - Selecting model(s) and data set

Above we obtained eight altered data sets. To have an idea, below I am going to make a simple cross-comparison of models and data sets. For the sake of simplicity, I will use the models with their default values.

```python
models = {'RFC': RandomForestClassifier(), 'ADA': AdaBoostClassifier(),
          'KNN':KNeighborsClassifier(), 'XTC': ExtraTreesClassifier(), 
          'GBS': GradientBoostingClassifier(), 'SVC': SVC(), 'GNB': GaussianNB()}

df_sets = {'glass': glass, 'glass_outs_dropped': glass_outs_dropped, 'glass_oc_scaled': glass_oc_scaled, 'glass_all_scaled': glass_all_scaled, 
           'glass_norm': glass_norm, 'gls_ots_drp_norm': gls_ots_drp_norm, 'gls_oc_scl_norm': gls_oc_scl_norm, 'gls_all_scl_norm': gls_all_scl_norm}

for model in models:
    for dataSet in df_sets:
        
        # Train test split the altered data sets.
        X = df_sets[dataSet].copy() # original data frame is copied to another data frame to avoid data loss
        y = X.pop('Type') # Return item and drop from frame. Recorded into target variable as series object
    
        from sklearn.model_selection import train_test_split
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
        
        # Apply 10 fold cross-validation. 
        cv_results = cross_val_score(models[model], X_train, y_train, cv=10, scoring = 'accuracy', n_jobs=-1)
        
        print('Accuracy score of', model, 'model with data set', dataSet, 'is %.2f' %(np.mean(cv_results)), '(mean values)')
```
